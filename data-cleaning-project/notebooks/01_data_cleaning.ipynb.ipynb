```python
import pandas as pd
```

```python
df=pd.read_csv("candlestick_chart.csv")
```

```python
df["Date"]=pd.to_datetime(df["Date"])
```

```python
df["Last Price"].head()
```

```python
df["Last Price"] =pd.to_numeric(df["Last Price"], errors="coerce")
```

```python
df["Last Price"].isna().sum()
```

```python
df["Last Price"].unique()[:20]
```

```python
df["LastPrice_imputed"] = df["Last Price"].isna()
```

```python
df["Last Price"] = df["Last Price"].fillna(df["Close Price"])
```

```python
df["Last Price"].isna().sum()
```

```python
df["Symbol"] = df["Symbol"].astype("category")
```

```python
df["Series"] = df["Series"].astype("category")
```

```python
df["Year"] = df["Date"].dt.year
```

```python
df=df.drop(columns=["Zear"])
```

```python
df["Date"] = df["Date"].dt.strftime("%d-%m")
```
